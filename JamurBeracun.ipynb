# Soal 1 - Jamur Beracun üçÑ

```python
dfjamur = pd.read_csv('mushrooms.csv')
dfjamur.head()
```

### Preparing Dataset

```python
dfjamur.isnull().sum()
```

```python
dfjamur.info()
```

```python
# Missing value : setelah di cek pada dataset yang ada terdapat icon "?"  pada stalk root yang mengindikasikan missing values

persenmisval = len(dfjamur[dfjamur['stalk-root']=='?'])/len(dfjamur) * 100
print(f'Persentase missing value pada stalk root = {persenmisval}%' )

# karena dianggap besar maka untuk keperluan model ini feature stalk root tidak disertakan dalam model
dfjamur = dfjamur.drop(columns = 'stalk-root')
dfjamur.head()
```

```python
# Handling Text Data (Categorical Value) : with Target = Classes
dffeaturex = pd.DataFrame()
dffeaturey = dfjamur['class'].apply(lambda x : 1 if x == 'e' else 0 ) # 1 Edible , 0 Poisonus

for i in range(1, len(dfjamur.columns)):
    dfdummy = pd.get_dummies(dfjamur[dfjamur.columns[i]], prefix = i)
    dffeaturex = pd.concat([dffeaturex,dfdummy], axis='columns')

dffeaturex
```

```python
# Split Dataset
from sklearn.model_selection import train_test_split

xtr, xts, ytr, yts = train_test_split(dffeaturex,dffeaturey, test_size = 0.2)
dtrain = pd.concat([xtr,ytr], axis=1)
dtes = pd.concat([xts,yts], axis=1)
```

### Model : LogisticRegression()

```python
from sklearn.linear_model import LogisticRegression
modeljamur = LogisticRegression()
```

```python
modeljamur.fit(xtr,ytr)
```

```python
yp = modeljamur.predict(xts)
yp
```

```python
# Score Model Awal

print(modeljamur.score(xtr,ytr))
print(modeljamur.score(xts,yts))
```

### Post-Model

#### Evaluation Metrics

```python
# Confusion Matrix
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(yts,yp, labels= [1,0])
dfcm = pd.DataFrame(
    cm, columns=['Pred 1','Pred 0'], index= ['Actual 1', 'Actual 0']
)
dfcm
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import recall_score
from sklearn.metrics import precision_score
from sklearn.metrics import f1_score
from sklearn.metrics import balanced_accuracy_score
```

```python
# pengecekan apakah data imbalanced atau balanced

print(len(dfjamur[dfjamur['class'] == 'p']))
print(len(dfjamur[dfjamur['class'] == 'e']))

print(len(ytr[ytr ==  1]))
print(len(ytr[ytr ==  0]))

print(len(yts[yts ==  1]))
print(len(yts[yts ==  0]))

# dari data yang ada disimpulkan bahwa balance :digunakan accuracy score dan error rate
```

```python
print('Akurasi = ', accuracy_score(yts,yp))
print('Error Rate = ', 1 - accuracy_score(yts,yp))
print('Precision(+) = ', precision_score(yts,yp, pos_label = 1))
print('Precision(-) = ', precision_score(yts,yp, pos_label = 0))
print('Recall(+) = ', recall_score(yts,yp, pos_label = 1))
print('Recall(-) = ', recall_score(yts,yp, pos_label = 0))
print('F1 Score(+) = ', f1_score(yts,yp, pos_label = 1))
print('F1 Score(-) = ', f1_score(yts,yp, pos_label = 0))
```

```python
# ROC and AUC

from sklearn.metrics import roc_curve, roc_auc_score

fpr, tpr, thres = roc_curve(yts, modeljamur.predict_proba(xts)[:,1])
roc_auc_score(yts,yp )
plt.plot(fpr, tpr)
plt.plot([0,1], [0,1], 'r:')
plt.xlabel('False Positive rate')
plt.ylabel('True Positive rate')
plt.show()
```

```python
# Luas AUC
roc_auc_score(yts,yp)
```

### Cross Validation

```python
from sklearn.model_selection import KFold
k = KFold(n_splits = 10)
```

```python
score = []

for train,validate in k.split(dtrain):
    xtrain = dtrain.iloc[train][list(xtr.columns.values)]
    ytrain = dtrain.iloc[train]['class']
    xvalid = dtrain.iloc[validate][list(xtr.columns.values)]
    yvalid = dtrain.iloc[validate]['class']
    

    model = LogisticRegression()
    model.fit(xtrain,ytrain)
    score.append(model.score(xvalid,yvalid))
```

```python
print(f'Score Model Logistic Regression = {np.mean(score)} ')
```
